### I am trying to have the top 10 countries with the highest GDP per year 'gdp_for_year ($)' column as well as the suicide rate per 100k, or the 'suicides/100k pop' column for pandas. Is there a way I could simultaneously write a command to generate both of these information for the 10 ten countries?

## Set up the programming enviroment

```
import pandas as pd
```

## Load the dataset...


```
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Thinkful/Suicides.csv',thousands=',')
```

```
df.head()
```

```
df.info()
```

```
countries_gdp = df.groupby(['country'])[' gdp_for_year ($) '].mean().reset_index()
countries_gdp
```

```
countries_gdp.sort_values(by=[' gdp_for_year ($) '],ascending=False).head(10)
```

```
countries_avg_suicide_rate = df.groupby(['country'])['suicides/100k pop'].mean().reset_index()
countries_avg_suicide_rate
```

```
countries_avg_suicide_rate.sort_values(by=['suicides/100k pop'],ascending=False).head(10)
```

### Merge 'countries_gdp' and 'countries_avg_suicide_rate' on the country column...

```
countries_gdp_suicide_rate = countries_gdp.merge(countries_avg_suicide_rate,how='left',on='country')
countries_gdp_suicide_rate
```
